```python
# 필요한 라이브러리 임포트
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
from tqdm import tqdm
import pandas as pd

# tqdm이 pandas의 apply와 함께 동작하도록 설정
tqdm.pandas()

# 1️⃣ 감정분석 모델 불러오기
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# 2️⃣ 파이프라인 생성
sentiment_pipeline = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

# 3️⃣ 감정 레이블 매핑 함수 정의
def map_sentiment(label):
    """
    모델의 출력 라벨 (예: '1 star', '3 stars', '5 stars')을
    부정=0, 중립=1, 긍정=2 로 매핑
    """
    stars = int(label.split()[0])
    if stars <= 2:
        return 0  # 부정
    elif stars == 3:
        return 1  # 중립
    else:
        return 2  # 긍정

# 4️⃣ tqdm 진행률 표시 + 감정분석 수행 (결측치 대비)
df_order_reviews_clean["sentiment_raw"] = df_order_reviews_clean["review_comment_message"].progress_apply(
    lambda x: sentiment_pipeline(str(x))[0]["label"] if pd.notnull(x) else "3 stars"
)

# 5️⃣ 레이블 매핑
df_order_reviews_clean["sentiment"] = df_order_reviews_clean["sentiment_raw"].progress_apply(map_sentiment)

# 6️⃣ 결과 확인
df_order_reviews_clean[["review_comment_message", "sentiment"]].head()
```

```python
from deep_translator import GoogleTranslator
from tqdm import tqdm
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import platform
import os

tqdm.pandas()

# ✅ 한글 폰트 자동 설정
def get_font_path():
    system = platform.system()
    if system == "Windows":
        return "C:/Windows/Fonts/malgun.ttf"  # 맑은 고딕
    elif system == "Darwin":  # macOS
        return "/System/Library/Fonts/AppleGothic.ttf"
    else:  # Linux (예: Colab, Ubuntu 등)
        return "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"

font_path = get_font_path()

# ✅ 번역 함수 (한국어)
def translate_to_ko(text):
    try:
        return GoogleTranslator(source='auto', target='ko').translate(text)
    except:
        return text  # 번역 실패 시 원문 유지

# ✅ 한국어로 번역된 컬럼 생성
df_order_reviews_clean["review_comment_ko"] = df_order_reviews_clean["review_comment_message"].progress_apply(
    lambda x: translate_to_ko(str(x)) if pd.notnull(x) else ""
)

# ✅ 감정별 텍스트 합치기
sentiment_texts = {
    0: " ".join(df_order_reviews_clean.loc[df_order_reviews_clean["sentiment"] == 0, "review_comment_ko"].dropna()),
    1: " ".join(df_order_reviews_clean.loc[df_order_reviews_clean["sentiment"] == 1, "review_comment_ko"].dropna()),
    2: " ".join(df_order_reviews_clean.loc[df_order_reviews_clean["sentiment"] == 2, "review_comment_ko"].dropna())
}

# ✅ 워드클라우드 색상 설정
colors = {0: "Reds", 1: "Blues", 2: "Greens"}

# ✅ 워드클라우드 시각화
for sentiment, text in sentiment_texts.items():
    plt.figure(figsize=(8, 6))
    wc = WordCloud(
        font_path=font_path,       # 한글 폰트 지정 (중요!)
        width=800,
        height=600,
        background_color="white",
        colormap=colors[sentiment],
        max_words=200
    ).generate(text)
    
    sentiment_name = {0: "부정", 1: "중립", 2: "긍정"}[sentiment]
    plt.title(f"감정 {sentiment} ({sentiment_name}) 워드클라우드", fontsize=18)
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.show()
```
