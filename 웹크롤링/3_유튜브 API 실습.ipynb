- ì—¬ëŸ¬ ì˜ìƒì— ëŒ€í•œ ìƒì„¸ ì •ë³´ ëª¨ë‘ ì–»ê¸°

```python

# ë™ì˜ìƒ ìƒì„¸ì •ë³´ API ì—”ë“œí¬ì¸íŠ¸
video_url = "https://www.googleapis.com/youtube/v3/videos"

# ìƒì„¸ì •ë³´ë¥¼ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸
stats_list = []

# ê° ë™ì˜ìƒì˜ ìƒì„¸ ì •ë³´ ìˆ˜ì§‘
for video_id in search_df['video_id']:
    # API ë§¤ê°œë³€ìˆ˜ ì„¤ì •
    video_params = {
        'key': API_KEY,
        'id': video_id,
        'part': 'statistics'
    }
    
    # API í˜¸ì¶œ
    video_response = requests.get(video_url, params=video_params)
    
    if video_response.status_code == 200:
        video_data = video_response.json()
        if video_data['items']:
            stats = video_data['items'][0]['statistics']
            stats_list.append({
                'video_id': video_id,
                'view_count': int(stats.get('viewCount', 0)),
                'like_count': int(stats.get('likeCount', 0)),
                'comment_count': int(stats.get('commentCount', 0))
            })
            print(f"ID: {video_id} ìƒì„¸ì •ë³´ ìˆ˜ì§‘ ì™„ë£Œ")

# ìƒì„¸ì •ë³´ ë°ì´í„°í”„ë ˆì„ ìƒì„±
stats_df = pd.DataFrame(stats_list)
print("\nìƒì„¸ì •ë³´ ë°ì´í„°:")
display(stats_df)
```

### 3. ì˜ìƒ ê²€ìƒ‰ ì •ë³´ì™€ ìƒì„¸ ì •ë³´ ë°ì´í„° ë³‘í•© ë° ì €ì¥

```python
# video_idë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë‘ ë°ì´í„°í”„ë ˆì„ ë³‘í•©
final_df = pd.merge(search_df, stats_df, on='video_id')

# URL ì»¬ëŸ¼ ì¶”ê°€
final_df['video_url'] = 'https://www.youtube.com/watch?v=' + final_df['video_id']

print("\nìµœì¢… ë°ì´í„°:")
display(final_df)

# ë°ì´í„° ì €ì¥
json_filename = f'youtube_results_{search_term}_{len(final_df)}ê°œ.json'

# DataFrameì„ JSONìœ¼ë¡œ ë³€í™˜í•˜ì—¬ ì €ì¥
# - orient='records'ëŠ” ë°ì´í„°ë¥¼ ë¦¬ìŠ¤íŠ¸ í˜•íƒœì˜ JSONìœ¼ë¡œ ì €ì¥
# - force_ascii=FalseëŠ” í•œê¸€ ë“±ì˜ ìœ ë‹ˆì½”ë“œ ë¬¸ìë¥¼ ê·¸ëŒ€ë¡œ ì €ì¥
# - indent=2ëŠ” JSON íŒŒì¼ì„ ë³´ê¸° ì¢‹ê²Œ ë“¤ì—¬ì“°ê¸°
final_df.to_json(json_filename, orient='records', force_ascii=False, indent=2)
```
